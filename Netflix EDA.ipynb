## Q. 2) In which year the highest number of TV Shows & Movies were released? Show withBar Graph

```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(10, 5))
Mdata = data[data['type']=='Movie']['release_year'].value_counts().nlargest(10)
Mdata = pd.DataFrame(Mdata)
TVdata = data[data['type']=='TV Show']['release_year'].value_counts().nlargest(10)
TVdata
Mdata = Mdata.reset_index()
sns.barplot(x = 'index',y = 'release_year' , data = Mdata, ax = axes[0])
axes[0].set_title('Movie')
axes[0].set_xlabel('year')
axes[0].set_ylabel('Count')

TVdata = TVdata.reset_index()
sns.barplot(x = 'index',y = 'release_year' , data = TVdata,ax = axes[1])
axes[1].set_title('TV Show')
axes[1].set_xlabel('TV Show')
axes[1].set_ylabel('Count')
plt.tight_layout()
plt.show()
```

conclusion: in 2018 the highest number of TV Shows & Movies were released

## Q. 10) What is the maximum duration of a Movie/Show on Netflix ?

```python
movies = data[data['type']=='Movie']
movies['duration'].fillna('0 min',inplace = True)
df = movies['duration'].str.split(' ', expand=True)[0]
df = df.astype(int)
df
df.sort_values(ascending = True)
```

```python
tvshow = data[data['type']=='TV Show']
df2 = tvshow['duration'].str.split(' ',expand = True)[0]
df2.isna().any()
df2.astype(int)
df2.sort_values(ascending = True)
```

conclusion: Max duration of Movies is 312 minutes
    & 
Max duration of TV show is 9 Series
